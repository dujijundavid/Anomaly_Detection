## Anomaly_Data Class object

```python
class Anomaly_Data:
    def __init__(self, df,time_col):
        self.df=df
        self.time_col = time_col 
    
    def get_client_and_timestamp(self):
        self.client = self.df.client_id
        self.time_col= self.df.stats_timestamp
    
        
    def find_label_df(self):
        filter_col = [col for col in self.df if col.startswith('label')]
        
        self.label_df=self.df[filter_col]
        self.nrow, self.ncol= label_df.shape
        
    def sparse_to_dense(self):
        self.anomaly_dic={}
        for i in range(self.nrow):
            for j in range(self.ncol):

                if(self.label_df.iloc[i,j] == 0):
                    pass
                else:

                    self.anomaly_dic[str(i)+"_"+str(j)] = self.time_col[i][:16]
                     
                    # Map i to Device, map j to Anomaly_Feature
                    # Store timestamp
```

```python
anomaly_obj = Anomaly_Data(df,df.stats_timestamp)
anomaly_obj.get_client_and_timestamp()
anomaly_obj.find_label_df()

anomaly_obj.sparse_to_dense()
```
